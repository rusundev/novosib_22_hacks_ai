```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score, r2_score, precision_score
```

```python
IN_DIR = '/kaggle/input/novosib-in/'
```

```python
df_train = pd.read_csv(IN_DIR + 'train_dataset_train.csv')
```

```python
df_test = pd.read_csv(IN_DIR + 'test_dataset_test.csv')
```

```python
#Отделение меток от данных

X = df_train[['Easting', 'Northing', 'Height', 'Reflectance']]
y = df_train[['Class']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
get_ipython().system('pip install -U lightautoml')
```

```python
from lightautoml.automl.presets.tabular_presets import TabularAutoML
from lightautoml.tasks import Task
```

```python
automl = TabularAutoML(task = Task('multiclass', metric = 'crossentropy'))
oof_pred = automl.fit_predict(df_train,  roles = {'target': 'Class', 'drop': ['id']}, verbose = 2)
test_pred = automl.predict(df_test)
```

```python
result = recall_score(df_train['Class'], test_pred, average='macro', zero_division=True)

print("Recall score",result)
```

```python
pd.DataFrame({'id': df_test.id,
              'Class': np.squeeze(test_pred.astype(int))}).to_csv(
    '/kaggle/working/light_auto_ml_solution.csv', sep=',', index=None)
```
